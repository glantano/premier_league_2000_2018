# Editar conjunto de datos

```python
#cambiar los datos de la columna FTR = Full Time Result
df['FTR'] = np.where(df['FTR'] == 'H','HOME',  #aquellos datos de la columna FTR tengan H, se cambie a HOME
            np.where(df['FTR'] == 'A','AWAY', 'DRAW')) #aquellos datos de la columna FTR tengan A, se cambie a AWAY, sino DRAW
```

```python
#debido que no existe la columna 'Winner' se crea esta nueva columna
df['Winner'] = np.where(df['FTR'] == 'HOME',df['HomeTeam'], #aquellos que en la columna 'FTR' diga Home, va el nombre del equipo local a la columna
              np.where(df['FTR'] == 'AWAY', df['AwayTeam'], "Draw"))  #aquellos que en la columna 'FTR' diga Away, va el nombre del equipo visitante a la columna
```

```python
#crear columna con el resultado del equipo que ganó
df['Result'] = np.where(df['FTR'] == 'HOME','Home Team Win',
                np.where(df['FTR'] == 'AWAY', 'Away Team Win', "Draw"))
```

```python
#para extraer el año de cada partido y agregarlo a una nueva columna de Year
df['Year']=pd.DatetimeIndex(df['Date']).year
```

```python
# Crear una columna con el total de goles y la diferencia de goles
df['TotalGoal'] = df['FTHG'] + df['FTAG']
df['GoalDif']= df['FTHG']- df['FTAG']
```

# Construyendo estadistica de equipos

```python
# filtra el DataFrame df, seleccionando solo las filas donde el valor en la columna 'FTR' no sea igual a 'DRAW'.
#Esto implica que solo se están considerando partidos que no terminaron en empate.
df2 = df[df.FTR != 'DRAW']
#Aquí se agrupan los datos en el DataFrame filtrado (df2) según la columna 'Winner'
#Se cuenta cuántas veces aparece cada equipo como ganador.
#con reset_index(name = "Count") se utiliza para reiniciar el indice del df2 y asignale nombre count con el recuento de victorias
avgWinning = df2.groupby('Winner')['Winner'].count().reset_index(name = "Count")

#se renombra la columnas especificadas, con inplace true es para que lo haga en el dataframe avgWinning
avgWinning.rename(columns = {'Winner':'Team', 'Count':'Victories'}, inplace=True)
```

```python
#ver como quedó el df avgWinning
avgWinning
```

```python
#construyendo dataset con estadistica de cada equipo
#con groupby agrupa los datos de HomeTeam (equipos)
#con .agg('sum') sumara los HST (tiros al arco)
#con reset:index se reestable que el nuevo dataframe su columna de la suma de HST se llamarpa Home Shots on Target

teamStatistics = df.groupby('HomeTeam')['HST'].agg('sum').reset_index(name = 'Home Shots on Target')

#de esta nueva dataframe llamado "teamStatistics" se cambiarpa el nombre HomeTeam a Team
teamStatistics = teamStatistics.rename(columns={"HomeTeam": "Team"})

#crear tabla AST (away shot target) con el total de tiro al arco del equipo visitanta
AST = df.groupby('AwayTeam')['AST'].agg('sum').reset_index(name = 'Away Shots on Target')

#Agregar la tabla AST al dataframe de TeamStatistics.
#como en la tabla teamStatistic no existe la columna 'Away Shots on Target', python lo agrega y le pone los valores de AST[]
teamStatistics['Away Shots on Target'] = AST['Away Shots on Target']

#agregar columna de total de tiros
teamStatistics['Total Shots on Target'] = teamStatistics['Home Shots on Target'] + teamStatistics['Away Shots on Target']

```

```python
#creación tabla de totales de goles del local
HG = df.groupby('HomeTeam')['FTHG'].agg('sum').reset_index(name = 'Home Goal')

#se agrega el total de goles en la tabla de teamStatistic
teamStatistics['Home Goal'] = HG['Home Goal']

#creación tabla de totales de goles de la visita
AG = df.groupby('AwayTeam')['FTAG'].agg('sum').reset_index(name = 'Away Goal')

#se agrega el total de goles en la tabla de teamStatistic
teamStatistics['Away Goal'] = AG['Away Goal']
```

```python
#creación tabla de goles en medio tiempo para el local
HHG = df.groupby('HomeTeam')['HTHG'].agg('sum').reset_index(name = 'Half Home Goal')

#agregar tabla de goles de medio tiempo del local a la tabla teamStatistic
teamStatistics['Half Home Goal'] = HHG['Half Home Goal']

#creacion tabla de goles en medio tiempo para la visita
HAG = df.groupby('AwayTeam')['HTAG'].agg('sum').reset_index(name = 'Half Away Goal')

#agregar tabla de goles de medio tiempo de la visita a la tabla teamStatistic
teamStatistics['Half Away Goal'] = HAG['Half Away Goal']
```

```python
#goles que te anotaron jugando de local
HGC = df.groupby('HomeTeam')['FTAG'].agg('sum').reset_index(name = 'Home Goal Conceded')

#agregar tabla a teamStatistic
teamStatistics['Home Goal Conceded'] = HGC['Home Goal Conceded']

#goles que te anotaron jugando de visita
AGC = df.groupby('AwayTeam')['FTHG'].agg('sum').reset_index(name = 'Away Goal Conceded')

#agregar tabla a teamStatistic
teamStatistics["Away Goal Conceded"] = AGC['Away Goal Conceded']
```

```python
#agregar totales de goles realizados
teamStatistics['Total Goal'] = teamStatistics['Home Goal'] + teamStatistics['Away Goal']

#agregar totales de goles recibidos
teamStatistics['Total Goal Conceded'] = teamStatistics['Home Goal Conceded'] + teamStatistics['Away Goal Conceded']

#Eficacia de goles anotados
teamStatistics['Goal/ST'] = teamStatistics['Total Goal'] / teamStatistics['Total Shots on Target']
```

```python
#numero de participaciones de local
participations = df.groupby('HomeTeam')['HomeTeam'].count().reset_index(name = 'Count')

#debido que todos juegan de local y visita se multiplica por 2 para hacer los partidos total
participations['Count'] = participations['Count'] * 2

#se guarda el total de participaciones en una columna nueva 'Games' en la tabla teamStadistic
teamStatistics['Games'] = participations['Count']
```

```python
#RECORDAR QUE DF2 NO CONSIDERA EMPATE (DRAW)
#contar el numero de victoria por equipos
victories = df2.groupby('Winner')['Winner'].count().reset_index(name = 'Count')

#agregar el numero de victorias al teamStatistic
teamStatistics['Victories'] = victories['Count']

#porcentaje de victorias/juegos
teamStatistics['Wins/Games'] = teamStatistics['Victories'] / teamStatistics['Games']

#promedio de goles anotados por partidos
teamStatistics['avgGoals'] = teamStatistics['Total Goal'] / teamStatistics['Games']

#promedio de goles recibido por partidos
teamStatistics['avgGoals Conceded'] = teamStatistics['Total Goal Conceded'] / teamStatistics['Games']
```

```python
#Total de faltas cometidas jugando de local
HF = df.groupby('HomeTeam')['HF'].agg('sum').reset_index(name = 'Count')

#total de falta cometidas jugando de visita
AF = df.groupby('AwayTeam')['AF'].agg('sum').reset_index(name = 'Count')

#se agrega el total de falta cometidad de local y visita a la tabla teamStatistic
teamStatistics['Home Fouls'] = HF['Count']
teamStatistics['Away Fouls'] = AF['Count']

#se crea tambien el total de falta cometidas
teamStatistics['Total Fouls'] = teamStatistics['Home Fouls'] + teamStatistics['Away Fouls']
```

```python
#total de tarjetas amarilla jugando de local
HY = df.groupby('HomeTeam')['HY'].agg('sum').reset_index(name = 'Count')

#total de tarjetas amarilla jugando de visita
AY = df.groupby('AwayTeam')['AY'].agg('sum').reset_index(name = 'Count')

#se agrega el total de tarjetas amarillas recibidas de local y visita a la tabla teamStatistic
teamStatistics['Home Yellow Cards'] = HY['Count']
teamStatistics['Away Yellow Cards'] = AY['Count']

#se crea tambien el total de tarjetas amarillas
teamStatistics['Total Yellow Cards'] = teamStatistics['Home Yellow Cards'] + teamStatistics['Away Yellow Cards']
```

```python
#total de tarjetas roja jugando de local
HR = df.groupby('HomeTeam')['HR'].agg('sum').reset_index(name = 'Count')

#toal de tarjetas rojas jugando de visita
AR = df.groupby('AwayTeam')['AR'].agg('sum').reset_index(name = 'Count')

#se agrega el total de tarjetas rojas recibidas de local y visita a la tabla teamStatistic
teamStatistics['Home Red Cards'] = HR['Count']
teamStatistics['Away Red Cards'] = AR['Count']

#se crea tambien el total de tarjetas rojas
teamStatistics["Total Red Cards"] = teamStatistics["Home Red Cards"] + teamStatistics["Away Red Cards"]

#promedio de cartar rebicidas por faltas
teamStatistics['Cards/Fouls'] = (teamStatistics['Total Yellow Cards'] + teamStatistics['Total Red Cards'] ) / teamStatistics['Total Fouls']
```
